```python
import pyodbc
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

driver = 'SQL Server'
server = 'MARCO'
db1 = 'AI_Eafit'
tcon = 'yes'
uname = 'aieafit'
pword = 'marANT1502'

cnxn = pyodbc.connect(driver='{SQL Server}', host=server, database=db1,
                      trusted_connection=tcon, user=uname, password=pword)
```

```python
query_compras = "SELECT Id_Producto, Numero_Orden_Compra FROM dbo.T_Fact_compra"
df_compras = pd.read_sql(query_compras, cnxn)
df_compras = df_compras.astype('str')
df_compras.head()
```

```python
df_gr = df_compras.groupby('Numero_Orden_Compra').agg(lambda x: list(x))
df_gr.head()
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(list_compras).transform(list_compras)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
apriori_compras = apriori(df, min_support=0.01)
```

```python
rules = association_rules(apriori_compras, metric="lift", min_threshold=1)
rules
```

```python
rules[ (rules['lift'] >= 20) &
       (rules['confidence'] >= 0.9) ]
```
